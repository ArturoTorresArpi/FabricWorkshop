### 1. Extract Data

```python
# Read data from Lakehouse's files
df_name_basics = spark.read.csv("Files/Fabric_Workshop/name.basics.tsv", sep='\t', header=True, inferSchema=True)
df_title_basics = spark.read.csv("Files/Fabric_Workshop/title.basics.tsv", sep='\t', header=True, inferSchema=True)
df_title_ratings = spark.read.csv("Files/Fabric_Workshop/title.ratings.tsv", sep='\t', header=True, inferSchema=True)
```

### 2. Load Data

```python
# Write data into Lakehouse's Tables
df_name_basics.write.format('delta').mode('overwrite').option('overwriteSchema', 'true').save('Tables/IMDb_People')
df_title_basics.write.format('delta').mode('overwrite').option('overwriteSchema', 'true').save('Tables/IMDb_Titles')
df_title_ratings.write.format('delta').mode('overwrite').option('overwriteSchema', 'true').save('Tables/IMDb_Titles_Ratings')
```
